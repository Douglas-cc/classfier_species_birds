- Remover alguns caracteres da coluna arquivos

```python
list_strings = ["xc", ".flac"]

for string in list_strings: 
    df_mir["arquivo"] = remove_caracter(df_mir.arquivo, string)
    
df_mir["arquivo"] = df_mir["arquivo"].astype(int)    
```
